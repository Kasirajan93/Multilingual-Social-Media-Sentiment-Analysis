```python
import pandas as pd
import re
```

```python
df = pd.read_csv("../data/processed/multilingual_sentiment.csv")

df.head()
```

```python
print(df.shape)
```

```python
df["cleaned_text"] = df["text"]
```

```python
df["cleaned_text"] = df["cleaned_text"].apply(
    lambda x: re.sub(r"http\S+|www\S+", "", str(x))
)
```

```python
df[["text","cleaned_text"]].head()
```

```python
df[["text","cleaned_text"]].head()
```

```python
df["cleaned_text"] = df["cleaned_text"].str.replace(
    "#",
    "",
    regex=False
)
```

```python
df["cleaned_text"] = df["cleaned_text"].apply(
    lambda x: re.sub(r"\d+", "", str(x))
)
```

```python
df["cleaned_text"] = df["cleaned_text"].apply(
    lambda x: re.sub(
        r"[^\w\s\u0B80-\u0BFF\u0900-\u097F]",
        "",
        str(x)
    )
)
```

```python
df["cleaned_text"] = df["cleaned_text"].str.lower()
```

```python
df["cleaned_text"] = df["cleaned_text"].apply(
    lambda x: " ".join(x.split())
)
```

```python
df = df[df["cleaned_text"].str.strip() != ""]
```

```python
df[
    ["text","cleaned_text"]
].sample(10)
```

```python
df.to_csv(
    "../data/processed/multilingual_sentiment_cleaned.csv",
    index=False
)

print("Saved successfully")
```
