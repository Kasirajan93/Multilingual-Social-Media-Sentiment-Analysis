```python
import pandas as pd

columns = [
    "target",
    "id",
    "date",
    "flag",
    "user",
    "text"
]

df = pd.read_csv(
    "../data/raw/english/english_sentiment140.csv",
    encoding="latin-1",
    names=columns
)

print("Shape:", df.shape)
print("\nColumns:")
print(df.columns)

print("\nTarget Counts:")
print(df["target"].value_counts())

df.head()
```

```python
# ======================================
# ENGLISH DATASET STANDARDIZATION
# ======================================

# Keep only required columns

english_df = df[["text", "target"]].copy()

# Convert labels

english_df["sentiment"] = english_df["target"].map({
    0: "Negative",
    4: "Positive"
})

# Add language column

english_df["language"] = "English"

# Keep final columns

english_df = english_df[
    ["text", "sentiment", "language"]
]

print(english_df.shape)

english_df.head()
```

```python
english_df.to_csv(
    "../data/processed/english_processed.csv",
    index=False
)
```

```python
hindi_df = pd.read_csv("../data/raw/hindi/hindi_sentiment_analysis_dedup.csv")

print(hindi_df.shape)
print(hindi_df.columns)

hindi_df.head()
```

```python
print(hindi_df["sentiment"].value_counts())
```

```python
# Keep only Positive and Negative

hindi_processed = hindi_df[
    hindi_df["sentiment"].isin(["positive", "negative"])
].copy()

# Rename column

hindi_processed = hindi_processed.rename(
    columns={
        "sentence": "text"
    }
)

# Capitalize labels

hindi_processed["sentiment"] = (
    hindi_processed["sentiment"]
    .str.capitalize()
)

# Add language column

hindi_processed["language"] = "Hindi"

# Final schema

hindi_processed = hindi_processed[
    ["text", "sentiment", "language"]
]

print(hindi_processed.shape)

print(
    hindi_processed["sentiment"].value_counts()
)

hindi_processed.head()
```

```python
hindi_processed.to_csv(
    "../data/processed/hindi_processed.csv",
    index=False
)
```
