```python
import pandas as pd
import numpy as np
import os

print("Dataset Merging Phase Started")
```

```python
english_df = pd.read_csv("../data/processed/english_processed.csv")
tamil_df = pd.read_csv("../data/processed/tamil_processed.csv")
hindi_df = pd.read_csv("../data/processed/hindi_processed.csv")

print("English:", english_df.shape)
print("Tamil:", tamil_df.shape)
print("Hindi:", hindi_df.shape)
```

```python
print(english_df.columns)
print(tamil_df.columns)
print(hindi_df.columns)
```

```python
language_summary = pd.DataFrame({
    "Language": ["English", "Tamil", "Hindi"],
    "Rows": [
        len(english_df),
        len(tamil_df),
        len(hindi_df)
    ]
})

language_summary
```

```python
english_sample = english_df.sample(
    n=20000,
    random_state=42
)

print(english_sample.shape)
```

```python
merged_df = pd.concat(
    [
        english_sample,
        tamil_df,
        hindi_df
    ],
    ignore_index=True
)

print(merged_df.shape)
```

```python
merged_df = merged_df.sample(
    frac=1,
    random_state=42
).reset_index(drop=True)

merged_df.head()
```

```python
print(
    merged_df["language"].value_counts()
)
```

```python
print(
    merged_df["sentiment"].value_counts()
)
```

```python
merged_df.to_csv(
    "../data/processed/multilingual_sentiment.csv",
    index=False
)

print("Dataset Saved Successfully")
```
